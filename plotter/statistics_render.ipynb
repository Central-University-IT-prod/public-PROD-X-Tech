```python
import faker
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Распределение по специализациям

```python
sns.set_theme()
sns.set_context('paper')
```

```python
specialization_data = {
    'Фронтендеры': 39,
    'Бекендеры': 45,
    'Мобильщики': 15
}

specialization_df = pd.DataFrame(
    list(specialization_data.items()),
    columns=['Специализация', 'Количество']
)

plot = sns.barplot(x='Специализация', y='Количество', data=specialization_df, color='#00ac8c')
```

```python
fig = plot.get_figure()
fig.savefig('plot_1.svg', format='svg')
```

Распределение с командой и без

```python
team_picked_data = {
    'Нашли команду': 56,
    'Без команды': 18
}

labels = team_picked_data.keys()
values = team_picked_data.values()

colors = ['#ff9999', '#00ac8c']

plt.pie(values, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

plt.axis('equal')

plt.savefig('plot_2.svg', format='svg')
```

# Сколько людей зарегистрировались на финал, сколько людей не зарегистрировались (им отправляем уведомление на почту)
